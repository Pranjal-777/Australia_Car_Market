```python
import pandas as pd

# Load the dataset
data = pd.read_csv('cars_info.csv')

# Display basic info
print(data.info())
print(data.head())
```

2. Average Price of Cars Across Brands


```python
# Group by brand and calculate average price
avg_price_by_brand = data.groupby('Brand')['Price'].mean()

# Find brands with the highest and lowest average prices
highest_avg_price = avg_price_by_brand.idxmax(), avg_price_by_brand.max()
lowest_avg_price = avg_price_by_brand.idxmin(), avg_price_by_brand.min()

print("Average Price by Brand:\n", avg_price_by_brand)
print("Highest Average Price:", highest_avg_price)
print("Lowest Average Price:", lowest_avg_price)
```

3. Effect of Car Condition on Price


```python
# Group by status and calculate average price
avg_price_by_status = data.groupby('Status')['Price'].mean()

print("Average Price by Condition:\n", avg_price_by_status)

# Visualize using a bar plot
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Status', y='Price', data=data)
plt.title("Average Price by Car Condition")
plt.show()
```

4. Correlation Between Mileage and Price


```python
# Calculate the correlation
correlation = data['Kilometers'].corr(data['Price'])
print("Correlation between Mileage and Price:", correlation)

# Visualize the relationship
sns.scatterplot(x='Kilometers', y='Price', data=data)
plt.title("Mileage vs Price")
plt.show()
```

5. Engine Capacity vs Price and Mileage


```python
# Correlation with price and mileage
cc_price_corr = data['CC'].corr(data['Price'])
cc_mileage_corr = data['CC'].corr(data['Kilometers'])

print("Correlation between Engine Capacity and Price:", cc_price_corr)
print("Correlation between Engine Capacity and Mileage:", cc_mileage_corr)

# Visualize relationships
sns.scatterplot(x='CC', y='Price', data=data)
plt.title("Engine Capacity vs Price")
plt.show()

sns.scatterplot(x='CC', y='Kilometers', data=data)
plt.title("Engine Capacity vs Mileage")
plt.show()
```
